```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/content/heart.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.shape
df.describe()
```

```python
df.target.value_counts()
```

```python
corr=df.corr()
plt.figure(figsize=(18,10))
sns.heatmap(corr,annot=True)
plt.show()
```

```python
sns.countplot(x='target',hue='sex',data=df,palette=['#1CA53B','red'])
plt.xticks([0,1],labels=['No Disease','Disease'])
plt.title('Target distribution')
plt.show()
```

```python
plt.figure(figsize=(18,10))
sns.countplot(x='age',hue='target',data=df,palette=['#1CA53B','red'])
plt.legend([' haven not Disease','have Disease'])
plt.title('heart diseases frequency for ages')
plt.xlabel('age')
plt.ylabel('frequency')
plt.show()
```

```python
plt.figure(figsize=(18,10))
sns.distplot(df.age[df['target']==0],bins=30,color='green',label='have not Disease')
sns.distplot(df.age[df['target']==1],bins=30,color='red',label='have Disease')
plt.legend()
plt.title('heart diseases frequency for ages')
plt.xlabel('age')
plt.ylabel('frequency')
plt.show
```
